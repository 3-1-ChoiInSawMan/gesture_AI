```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("./dataset/good/000.npy")
frame = data[0].reshape(44, 2)

uniq, counts = np.unique(frame, axis=0, return_counts=True)
print(list(zip(uniq, counts)))

# 많이 겹치는 점 확인
mask = ~(np.all(frame == frame[0], axis=1))  # 단순 예시
plt.scatter(frame[mask, 0], frame[mask, 1])
plt.gca().invert_yaxis()
plt.show()
```

```python
import numpy as np

data = np.load("./dataset/test/001.npy")
frame = data[0].reshape(44, 2)

uniq, counts = np.unique(frame, axis=0, return_counts=True)

print("unique 좌표 수:", len(uniq))
for u, c in zip(uniq, counts):
    print(u, c)
```

```python
import numpy as np

data = np.load("dataset/test/003.npy")
print(data.shape)

frame = data[0].reshape(44, 2)
uniq = np.unique(np.round(frame, 5), axis=0)
print("unique 좌표 수:", len(uniq))
```

```python
for i in [0, 10, 20, 30, 40, 50]:
    frame = data[i].reshape(44, 2)
    print(i, len(np.unique(np.round(frame, 5), axis=0)))
```
