```python
import asyncio
from pathlib import Path

import websockets

WS_URL = "ws://localhost:8000/cc_stt"
WAV_PATH = Path("..") / "test.wav"

async def main():
    async with websockets.connect(WS_URL) as ws:
        await ws.send(WAV_PATH.read_bytes())

        result = await asyncio.wait_for(ws.recv(), timeout=10)
        print(result)

await main()
```
