```python
import asyncio
import websockets

WS_URL = "ws://localhost:8000/stt"
WAV_PATH = "test.wav"

async def main():
    async with websockets.connect(WS_URL) as ws:
        with open(WAV_PATH, "rb") as f:
            wav_bytes = f.read()

        await ws.send(wav_bytes)

        result = await ws.recv()
        print(result)

asyncio.run(main())
```
